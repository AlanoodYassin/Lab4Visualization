```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
Anoodsdata = pd.read_csv('Wholesale customers data.csv')
from sklearn.preprocessing import normalize
data_scaled = normalize(Anoodsdata)
```

```python
data_scaled = pd.DataFrame(data_scaled, columns=Anoodsdata.columns)

import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))  

dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
```

```python
plt.figure(figsize=(10, 7))  

plt.title("AnoodsDendrogram")

z = shc.linkage(data_scaled, method='ward')

dend = shc.dendrogram( z )

plt.axhline(y=6, color='r', linestyle='--')

plt.show()
```
